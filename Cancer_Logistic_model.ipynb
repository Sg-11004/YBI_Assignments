```python
df = df.loc[:, ~df.columns.str.contains('Unnamed: 32')]
```

```python
df.columns
```

```python
x=df['diagnosis']
y=[['id','radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
```

```python

X = df.drop(['id', 'diagnosis'], axis=1)  # Features
y = df['diagnosis']  
```

```python
df.describe()
```

```python
df.info()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2529)
```

```python
df.shape,X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
model=LogisticRegression(max_iter=100000)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(f'accuracy score:{accuracy_score(y_test,y_pred)*100}')
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Assuming y_test are your true labels and y_pred are your model's predictions

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')

# Precision
precision = precision_score(y_test, y_pred, pos_label=1)
print(f'Precision: {precision}')

# Recall
recall = recall_score(y_test, y_pred, pos_label=1)
print(f'Recall: {recall}')

# F1 Score
f1 = f1_score(y_test, y_pred, pos_label=1)
print(f'F1 Score: {f1}')
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, roc_curve, auc

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(7, 7))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion matrix')
plt.colorbar()
plt.xticks([0, 1], ['Negative', 'Positive'], rotation=45)
plt.yticks([0, 1], ['Negative', 'Positive'])
plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
# ROC Curve
y_pred_proba = model.predict_proba(X_test)[::,1]
fpr, tpr, _ = roc_curve(y_test,  y_pred_proba)
auc = auc(fpr, tpr)
plt.figure(figsize=(7, 7))
plt.plot(fpr, tpr, label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.title('ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```
